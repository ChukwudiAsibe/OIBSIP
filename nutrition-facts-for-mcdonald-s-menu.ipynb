**EXPLORATORY DATA ANALYSIS**

**Inspiration**

How many calories does the average McDonald's value meal contain? How much do beverages, like soda or coffee, contribute to the overall caloric intake? Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value? What about ordering egg whites instead of whole eggs? What is the least number of items could you order from the menu to meet one day's nutritional requirements?

```python
print(f"The average Calories McDonald's value meal contains is: {round(nutrition.describe().Calories.iloc[1], 2)}cal")
```

```python
overall_calories = (nutrition.Calories + nutrition['Calories from Fat']).sum()

Beverages = nutrition[nutrition.Category == 'Beverages']
Beverages_calories = (Beverages.Calories + Beverages['Calories from Fat']).sum()
print(f"Beverages contribution to the overall caloric intake is: {round(Beverages_calories / overall_calories * 100, 2)}%")
```

```python
sandwich = nutrition[nutrition.Item.str.contains('sandwich', case=False)]
crispy = sandwich[sandwich.Item.str.contains('crispy', case=False)]
grilled_chicken = sandwich[sandwich.Item.str.contains('grilled chicken', case=False)]

sanwhich_overal_nutriton = sandwich.sum(axis=1, numeric_only=True).sum()
crispy_overal_nutriton = crispy.sum(axis=1, numeric_only=True).sum()
grilled_chicken_overal_nutriton = grilled_chicken.sum(axis=1, numeric_only=True).sum()

print(f"sandwich's nutritional value of crispy is: {round(crispy_overal_nutriton / sanwhich_overal_nutriton * 100, 2)}%")
print(f"sandwich's nutritional value of grilled chicken is: {round(grilled_chicken_overal_nutriton / sanwhich_overal_nutriton * 100, 2)}%")
```

```python
egg = nutrition[nutrition.Item.str.contains('egg', case=False)]
overall_eggs_nutrition = egg.sum(axis=1, numeric_only=True).sum()

egg_white = nutrition[nutrition.Item.str.contains('egg white', case=False)]
egg_white_overal_nutrition = egg_white.sum(axis=1, numeric_only=True).sum()

egg = egg[~egg.isin(egg_white)].dropna()
egg_overall_nutrition = egg.sum(axis=1, numeric_only=True).sum()

print(f"Nutritional value of egg is: {round(egg_overall_nutrition / overall_eggs_nutrition * 100, 2)}%")
print(f"Nutritional value of egg whites is: {round(egg_white_overal_nutrition / overall_eggs_nutrition * 100, 2)}%")
```
